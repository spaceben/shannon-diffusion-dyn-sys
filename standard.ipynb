```julia
using DynamicalSystems
using PyPlot
```

```julia

function plot_sm(k)
    cm = Systems.standardmap(k=k)
    for i = 1:10
        tr = trajectory(cm, 10000, rand(Float64,2) * 2pi)
        plot(tr[:, 1]/(2pi), tr[:, 2]/(2pi), lw = 0.0, marker = "o", ms = 0.1, alpha = 0.5);
    end
end



fig = figure("pyplot_subplot_column",figsize=(10,5))

subplot(131) # Create the 1st axis of a 3x1 array of axes
PyPlot.title("k = 0.03")
plot_sm(0.03)

ax = gca()
ax[:set_ylim]([0,1])
ax[:set_xlim]([0,1])
xlabel("x"); ylabel("y");

subplot(132)
PyPlot.title("k = k_crit ~ 0.6")
plot_sm((sqrt(5)-1)/2)

ax = gca()
ax[:set_ylim]([0,1])
ax[:set_xlim]([0,1])
xlabel("x");

subplot(133)
PyPlot.title("k = 3")
plot_sm(3)

xlabel("x")
ax = gca()
ax[:set_ylim]([0,1])
ax[:set_xlim]([0,1])

savefig("standard.pdf")
```
